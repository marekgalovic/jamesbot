```python
import json
```

```python
frames = json.load(open('../data/frames/frames.json'))
```

```python
act_types = {}
intent_types = set()

for dialog in frames:
    for turn in dialog['turns']:
        for act in turn['labels'].get('acts', []):
            if act['name'] in act_types:
                act_types[act['name']].extend(act['args'])
            else:
                act_types[act['name']] = act['args']
                
            if act['name'] == 'inform':
                for arg in act['args']:
                    if arg.get('key') == 'intent':
                        intent_types.add(arg.get('val'))
```

```python
for (name, args) in act_types.items():
    print(name)
    for arg in args[:10]:
        print(arg)
    print()
```
